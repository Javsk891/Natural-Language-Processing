```python
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

token_1 = word_tokenize('Mackbeth - Shakespiere.txt')
token_2 = word_tokenize('The Time Machine - H. G Wells.txt')
```

```python
print('Tokens del texto_1', token_1)
print('Tokens del texto_2', token_2)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier

# Textos de ejemplo
texto_Mackbeth = 'Mackbeth - Shakespiere.txt'
texto_MachineTime = 'The Time Machine - H. G Wells.txt'
# Crear corpus
corpus = [texto_Mackbeth, texto_MachineTime]

# Crear el vectorizador TF-IDF
vectorizer = TfidfVectorizer()

# Vectorizar el corpus
X = vectorizer.fit_transform(corpus)

# Etiquetas para cada texto
y = np.array([0, 1])  # 0 para texto1, 1 para texto2

# Crear el modelo KNN
knn_model = KNeighborsClassifier(n_neighbors=1)
knn_model.fit(X, y)

# Función para predecir la pertenencia de una palabra a un texto
def predict_text(word):
    word_vector = vectorizer.transform([word])
    prediction = knn_model.predict(word_vector)
    return prediction[0]

# Pruebas
word_to_check = 'machine time'
predicted_label = predict_text(word_to_check)

if predicted_label == 0:
    print(f'La palabra "{word_to_check}" pertenece al texto1.')
else:
    print(f'La palabra "{word_to_check}" pertenece al texto2.')
```
